```python
# convert to one-shots
trainYproc = np.zeros((trainY.shape[0], 2))
for i in range(trainY.shape[0]):
    if trainY[i,0] == 0:
        trainYproc[i,0] = 1
    elif trainY[i,0] == 1:
        trainYproc[i,1] = 1
    else:
        raise ValueError
trainY = trainYproc

testYproc = np.zeros((testY.shape[0], 2))
for i in range(testY.shape[0]):
    if testY[i,0] == 0:
        testYproc[i,0] = 1
    elif testY[i,0] == 1:
        testYproc[i,1] = 1
    else:
        raise ValueError
testY = testYproc
```

```python
# build neural network
import math
class NeuralNetwork:
    '''Represents a neural network. Provides forward and back prop functions.'''
    
    def __init__(self, inputSize, hiddenSize, outputSize):
        self.W1 = np.matrix(np.zeros((hiddenSize, inputSize+1)))
        self.W2 = np.matrix(np.zeros((outputSize, hiddenSize+1)))
        self.sigmoid = np.vectorize(lambda x: 1/(1+math.exp(-x)))
        self.prevdW1 = np.zeros(self.W1.shape)
        self.prevdW2 = np.zeros(self.W2.shape)
        
    def forwardProp(self, X):
        self.z1 = self.addBias(X) * self.W1.T
        self.hidden = self.sigmoid(self.z1)
        self.z2 = self.addBias(self.hidden) * self.W2.T
        self.output = self.sigmoid(self.z2)
        return self.output
    
    def addBias(self, X):
        return np.c_[np.ones((X.shape[0], 1)), X]
    
    def backwardProp(self, X, y):
        self.forwardProp(X)
        X = self.addBias(X)
        m = X.shape[0]
        L2 = self.output.T
        L1 = self.hidden.T
        d = np.multiply(L2 - y.T, np.multiply(L2, 1-L2))
        self.dW2 = 1/m * d * self.addBias(L1.T)
        self.dW1 = 1/m * np.multiply((self.W2.T * d)[1:,:] , np.multiply(L1, 1 - L1)) * X
        
    def updateWeights(self, alpha=0.5, beta=0.01):
        '''Call backprop before this method'''
        self.W2 -= alpha * self.dW2 + beta * self.prevdW2
        self.W1 -= alpha * self.dW1 + beta * self.prevdW1
        self.prevdW2 = self.dW2
        self.prevdW1 = self.dW1
    
    def cost(self, X, y):
        self.forwardProp(X)
        return 1/(2 * X.shape[0]) * np.sum((self.output - y).T * (self.output - y))
        
nn = NeuralNetwork(testX.shape[1], testX.shape[1]+5, 2)
```
